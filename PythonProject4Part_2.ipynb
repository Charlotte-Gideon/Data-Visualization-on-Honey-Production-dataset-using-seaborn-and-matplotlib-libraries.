## 3. Display the percentage distribution of the data in each year using the pie chart.

```python
##############    Percentage distribution of the data in each year    ##############
df1=df['year'].value_counts()*100/df.shape[0]
df1
plt.figure(figsize=(16,8))
plt.pie(df1,labels=df1.index,autopct='%1.1f%%')
```

## 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
sns.regplot(df,x='numcol',y='prodvalue',ci=0)
```

* From the obtained graph we can infer that the trendline in not flat and has a fine slope ,therefore the columns 'numcol' and 'prodvalue' are dependant on each other and may have positive correlation that is they are directly proportional to each other. 

## 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
sns.boxplot(df,x='year',y='prodvalue')
```

* From the graph we can infer that over the years the prodvalue did not remain constant and fluctuated and the outliers of higher values are found around 2010 to 2012.

## 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year').

```python
sns.pairplot(df[['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year']],kind='reg',diag_kind='hist',hue='year')
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue').

```python
#################    Correlation metric   #################
df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr()
```

```python
sns.heatmap(df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr(),annot=True)
```

* From the above graph we can infer that there high positive correlation between (numcol and totalprod),(numcol and stocks),(numcol and prodvalue),(totalprod and stocks),(totalprod and prodvalue),(stocks and prodvalue)
* From this we can infer that the dataset has high multicolinearity as there varibles which has high magnitude of correlation with other variables that is there are variables which depends on the other variables.
